```python
import numpy as np
import os

import matplotlib.pyplot as plt
import seaborn as sns

from matplotlib import cm
%matplotlib inline
sns.set_theme()
sns.set(font_scale=1.25)
```

# Linear encoding

## Plot for each epoch

```python
min_layers, max_layers = 1,4
min_sublayers, max_sublayers = 1,2

cmap = cm.get_cmap('plasma', (max_layers-min_layers+1)*(max_sublayers-min_sublayers+1)+1)

fig, ax = plt.subplots(figsize=(8,4))
# ax.set_ylim(0.05,.4)

#inset settings
axin_1 = ax.inset_axes([0.5, 0.5, 0.4, 0.45])
axin_1.set_ylim([0.055, 0.16])
axin_1.set_xlim([80, 100])
ax.indicate_inset_zoom(axin_1, edgecolor='grey',)#'black'
plt.setp(axin_1.spines.values(), color='black')

i = 0

for layers in range(min_layers, max_layers+1):
    for sublayers in range(min_sublayers, max_sublayers+1):

        dir_path = '..'

        data = dir_path+f'/results/linear/{layers}l-{sublayers}p'
        val_cost = np.load(data+'/val_cost_per_epoch.npy', )
        train_cost = np.load(data+'/train_cost_per_epoch.npy', )


        pl1, = ax.plot(train_cost,linestyle='dashed',# label=f'{layers} l, {sublayers}sl - Train', 
                       c=cmap(i),  )#alpha=0.5
        axin_1.plot(train_cost,linestyle='dashed',# label=f'{layers} l, {sublayers}sl - Train', 
                    c=cmap(i),  )#alpha=0.5
        pl2, = ax.plot(val_cost, label=f'{layers} l, {sublayers}sl',
                       c=cmap(i))
        axin_1.plot(val_cost, label=f'{layers} l, {sublayers}sl',
                    c=cmap(i))
        i+=1



ax.set_ylabel('MSE')
ax.set_xlabel('Epochs')
lgd = ax.legend(bbox_to_anchor=(.0, 1.3), ncol=4, loc='upper left')#
# ax.add_artist(lgd)
# lgd1 = ax.legend([pl1,pl2],['Train','Val'], loc='lower left',bbox_to_anchor=(0.1, 0.3))#
# ax.add_artist(lgd)
#creating a folder to save the plot
dir_path = '..'
data = dir_path+f'/plots/linear'
os.makedirs(data,  0o755,  exist_ok=True)

fig.savefig(data+'/compare_mse_per_epoch.pdf',
            bbox_extra_artists=(lgd,), bbox_inches='tight')
plt.show()
```
