```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set_theme()
```

## Benchmark

```python
df = pd.read_csv('./benchmark.csv', sep=';', index_col=0)
df.head()
```

```python
_, ax = plt.subplots(1, 1)

X = ['L'+x.split('-')[0] + 'S'+x.split('-')[1] for x in df.columns.values]
X_axis = np.arange(len(X))

y_linear = df.loc['Linear'].values
y_nonlinear = df.loc['Nonlinear'].values

ax.bar(X_axis - 0.2, y_linear, 0.4, label = 'Linear')
ax.bar(X_axis + 0.2, y_nonlinear, 0.4, label = 'Non Linear')

plt.xticks(X_axis, X)
ax.set_xlabel("Composition of Layers and Sublayers")
ax.set_ylabel("MSE")
ax.set_title("Hyperparameter tuning")
ax.legend()
plt.show()
```
